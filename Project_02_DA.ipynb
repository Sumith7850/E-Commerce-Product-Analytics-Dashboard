```python
import requests
import pandas as pd
url="https://fakestoreapi.com/products"
responces = requests.get(url)
data = responces.json()
df = pd.DataFrame(data)
df.head(3)
```

```python
df.description
```

```python
df['rating_rate']=df['rating'].apply(lambda x:x['rate'])
df['rating_count']=df['rating'].apply(lambda x:x['count'])
df=df.drop(columns=['rating'])
df.head()
```

```python
def to_camel_case(col):
    parts = col.split(" ")
    return parts[0]+' '.join(word.capitalize() for word in parts[1:])
df.columns=[to_camel_case(col) for col in df.columns]
print(df.columns)
```

```python
df['price']=df['price'].round(0)
df['price']=df['price'].astype(int)
```

```python
df['title']=df['title'].str.strip()
df['category']=df['category'].str.lower().str.strip()
df['description']=df['description'].str.strip()
```

```python
df['category'] = df['category'].replace({
    "men's clothing": "men",
    "women's clothing": "women",
    "jewelery": "jewelry"
})
```

```python
def price_category(price):
    if price <50:
        return "Low"
    elif price <150:
        return "medium"
    else:
        return "High"
    
df["price_Category"]=df["price"].apply(price_category)
```

```python
df.head()
```

```python
df['titleLength'] = df['title'].apply(len)
```

```python
df['Popularity_Score']=df["rating_rate"]*df["rating_count"]
```

```python
df=df.sort_values(by='rating_rate',ascending=False)
```

```python
df.head(3)
```

```python
print(df.groupby('category')['price'].mean())
```

```python
print(df.groupby('price_Category')['id'].count())
```

```python
df.to_csv("cleaned_products.csv", index=False)
```

```python
import pyodbc

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=Localhost\SQLDEV;"   # change this
    "DATABASE=ecommerce_db;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

print("✅ Connected to SQL Server!")
```

```python
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO products (
            id, title, price, category, description, image,
            rating_rate, rating_count,
            price_Category, Popularity_Score, titleLength
        )
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        row['id'],
        row['title'],
        row['price'],
        row['category'],
        row['description'],
        row['image'],
        row['rating_rate'],
        row['rating_count'],
        row['price_Category'],
        row['Popularity_Score'],
        row['titleLength']
    ))

conn.commit()
print("✅ Data inserted successfully!")
```
